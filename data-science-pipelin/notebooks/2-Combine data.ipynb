```python
# create.a big. dataframe
my_entire_dataframe =pd.DataFrame()
my_dataframes =[]
```

```python
csv_files = [x for x in os.listdir(DATA_DIR) if x.endswith(".csv")]#only take csv files

for fname in os.listdir(DATA_DIR):
    print(fname)
    csv_path = os.path.join(DATA_DIR, fname)
    this_df = pd.read_csv(csv_path)
    this_df['filename'] = fname
    year = fname.replace(".csv","")
    this_df['year'] = year
    print(this_df.head(n=1))
    my_dataframes.append(this_df)
```

```python
my_entire_dataframe = pd.concat(my_dataframes) #combine dataframes into one
```
